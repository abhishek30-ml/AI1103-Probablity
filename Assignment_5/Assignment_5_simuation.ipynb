```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100
m=45
t=15
x=[]
for i in range(m):
    a = np.random.uniform(0,t)
    x.append(a)
```

```python
theta = np.linspace(5,50,100)
logL = n*np.log(1/theta) - ((n-m)*t + np.sum(x))/theta
theory_theta = (((n-m)*t + np.sum(x))/n)*np.ones(np.size(theta))
```

```python
plt.plot(theta, logL)
plt.plot(theory_theta, logL)

plt.xlabel('theta')
plt.ylabel('Log Likelihood')
```
